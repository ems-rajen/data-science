3. ***panel :***
* it is a 3D container of data
* we can create it from ndarrays orm dict of DataFrames
* syntax : pandas.Panel(data, items, major_axis, minor_axis, dtype, copy)
* 3 axes are:
* * items − axis 0, each item corresponds to a DataFrame contained inside.
* * major_axis − axis 1, it is the index (rows) of each of the DataFrames.
* * minor_axis − axis 2, it is the columns of each of the DataFrames.

<b>Syntax:</b>

pandas.Panel(data, items, major_axis, minor_axis, dtype, copy)

* <b>example : </b>

<code>

(Creating a Panel)

print(p)

#### Pandas Array Copy vs View:
it works same as numpy's copy and view method<br>
***The Difference Between Copy and View -***<br>
<br>1. The main difference between a copy and a view of an array is that the copy is a new array, and the view is just a view of the original array.
<br>2. The copy owns the data and any changes made to the copy will not affect original array, and any changes made to the original array will not affect the copy.
<br>3. The view does not own the data and any changes made to the view will affect the original array, and any changes made to the original array will affect the view.

<b>1. view: </b> newdf1 is just a view of newdf, newdf is just like a pointer of newdf ,so whatever you'll perform then newdf will be change and if you modify original newdf then also change newdf1.

Both numpy.NaN and pandas.NaT are used to represent missing data, but they are intended for different types of data:

##### numpy.NaN (Not a Number):
Type: Floating-point number (float).
Usage: Represents missing numerical data.
Origin: Part of the IEEE 754 floating-point standard, used by NumPy.
Behavior: Operations with NaN typically result in NaN. For example, any arithmetic operation involving NaN will result in NaN.
##### pandas.NaT (Not a Time):
Type: Special constant used by pandas for missing datetime data.
Usage: Represents missing values in datetime columns.
Origin: Specific to pandas, designed to handle missing dates and times.
Behavior: Operations involving NaT with datetime functions or arithmetic operations typically result in NaT.
Here is an example demonstrating their usage and behavior:

```python
import numpy as np
import pandas as pd

# Creating NaN and NaT examples
nan_value = np.NaN
nat_value = pd.NaT

# Displaying their types and values
print(f'nan_value: {nan_value}, type: {type(nan_value)}')
print(f'nat_value: {nat_value}, type: {type(nat_value)}')

# Operations with NaN
print(f'np.NaN + 1: {np.NaN + 1}')  # Results in NaN

# Operations with NaT
print(f'pd.NaT + pd.Timedelta(days=1)')  # Results in NaT
```

* Use numpy.NaN for missing numerical or general data.
* Use pandas.NaT for missing datetime data.

#### Pandas - Fixing Wrong Data
<b>Wrong Data : </b>
"Wrong data" does not have to be "empty cells" or "wrong format", it can just be wrong, like if someone registered "199" instead of "1.99".
<br>
Sometimes you can spot wrong data by looking at the data set, because you have an expectation of what it should be.
<br>
If you take a look at our data set, you can see that in row 7, the duration is 450, but for all the other rows the duration is between 30 and 60.
<br>
It doesn't have to be wrong, but taking in consideration that this is the data set of someone's workout sessions, we conclude with the fact that this person did not work out in 450 minutes.

* ##### How can we fix wrong values, like the one for "Duration" in row 7?

##### Replacing Values
One way to fix wrong values is to replace them with something else.
<br>
In our example, it is most likely a typo, and the value should be "45" instead of "450", and we could just insert "45" in row 7:

```python
# Set "Duration" = 45 in row 7:

df.loc[7, 'Duration'] = 45
```

```python
df
```

<b>NOTE : </b> For small data sets you might be able to replace the wrong data one by one, but not for big data sets.
<br>
To replace wrong data for larger data sets you can create some rules, e.g. set some boundaries for legal values, and replace any values that are outside of the boundaries.

* Loop through all values in the "Duration" column.
* If the value is higher than 120, set it to 120:

```python
for x in df.index:
    if df.loc[x, "Duration"] > 120:
        df.loc[x, "Duration"] = 120
```

#### Removing Rows
Another way of handling wrong data is to remove the rows that contains wrong data.
<br>
This way you do not have to find out what to replace them with, and there is a good chance you do not need them to do your analyses.

* Delete rows where "Duration" is higher than 120:

```python
for x in df.index:
    if df.loc[x, "Duration"] > 120:
        df.drop(x, inplace = True)
```

***Note:*** The corr() method ignores "not numeric" columns.

### Result Explained
The Result of the corr() method is a table with a lot of numbers that represents how well the relationship is between two columns.
<br>
The number varies from -1 to 1.
<br>
1 means that there is a 1 to 1 relationship (a perfect correlation), and for this data set, each time a value went up in the first column, the other one went up as well.
<br>
0.9 is also a good relationship, and if you increase one value, the other will probably increase as well.
<br>
-0.9 would be just as good relationship as 0.9, but if you increase one value, the other will probably go down.
<br>
0.2 means NOT a good relationship, meaning that if one value goes up does not mean that the other will.
<br>
What is a good correlation? It depends on the use, but I think it is safe to say you have to have at least 0.6 (or -0.6) to call it a good correlation.
<br><br>
***Perfect Correlation:***<br>
We can see that "Duration" and "Duration" got the number 1.000000, which makes sense, each column always has a perfect relationship with itself.
<br><br>
***Good Correlation:***<br>
"Duration" and "Calories" got a 0.922721 correlation, which is a very good correlation, and we can predict that the longer you work out, the more calories you burn, and the other way around: if you burned a lot of calories, you probably had a long work out.
<br><br>
***Bad Correlation:***<br>
"Duration" and "Maxpulse" got a 0.009403 correlation, which is a very bad correlation, meaning that we can not predict the max pulse by just looking at the duration of the work out, and vice versa.
